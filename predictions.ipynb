```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import numpy as np
import pandas as pd
```

```python
model = load_model('model.h5')


## Load encoder and scaler
with open('label_encoder_gender.pkl', 'rb') as f:
    label_encoder_gender = pickle.load(f)
    
with open('onehot_encoder_geo.pkl', 'rb') as f:
    onehot_encoder_geo = pickle.load(f)
    
with open('scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)

```

```python
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
# applying transformations on input data
input_df = pd.DataFrame([input_data])
input_df['Gender'] = label_encoder_gender.transform([input_df['Gender']])
geo_encoded_input_df = pd.DataFrame(onehot_encoder_geo.transform([[input_data['Geography']]]), 
                                    columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
input_df = pd.concat([input_df.drop(columns=['Geography']), geo_encoded_input_df], axis=1)
input_scaled = scaler.transform(input_df)
```

```python
input_scaled
```

```python
## Prediction

prediction = model.predict(input_scaled)
prediction_proba = prediction[0][0]
```

```python
prediction_proba
```

```python
if prediction_proba > 0.5:
    print("The customer is likely to churn")
else:
    print("The customer is likely to not churn")
```
